```python
# swap case
def swap_case(s):
    swap = s.swapcase()
    return swap

message = 'How Are You?'
print(swap_case(message))
```

```python
# String Split and Join
def split_and_join(text):
    split = text.split()
    join = '-'.join(split)
    return join

text = 'part of a string'
print(split_and_join(text))
```

```python
# print_full_name function
def print_full_name(first, last):
    full_name = first + ' ' + last
    # full_name = f'{first} {last}'
    print(f'Hello {full_name}! You just start learning python.')

firstName = 'Tas'
lastName = 'Nim'
print_full_name(firstName, lastName)
```

```python
# data mutability 
# change the first . to _
# strings are immutable, list are mutable
filename = ['1.Raw Data.txt', '2.Reports.txt', '3.presentations.txt', '44.new.item.text']
for item in filename:
    new_item = item.replace('.', '_', 1)
    print(new_item)
```

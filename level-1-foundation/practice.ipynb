```python
# swap case
def swap_case(s):
    swap = s.swapcase()
    return swap

message = 'How Are You?'
print(swap_case(message))
```

```python
# String Split and Join
def split_and_join(text):
    split = text.split()
    join = '-'.join(split)
    return join

text = 'part of a string'
print(split_and_join(text))
```

```python
# generate new files and write in the file
contents = ['hello', 
            'Good morning', 
            'How are you?']

files = ['doc.txt', 'report.txt', 'presentation.txt']

for content, filename in zip(contents, files):
    # print(contents, filename)
    file = open(f'files/{filename}', 'w')
    file.write(content)
    file.close()

for content in files:
    file = open(f'files/{content}', 'r')
    file.read()

```

```python
# 1. generates multiple text files by iterating over the filenames list,
# 2. and writes the text Hello  inside each generated text file.

files = ['doc1.txt', 'report1.txt', 'presentation1.txt']
for filename in files:
    file = open(filename, 'w')
    file.write('Hello')
    file.close()
```
